## Normalization

### 1. MinMax Scaling
$$
x_i' = \frac{x_i - x_{\min}}{x_{\max} - x_{\min}}, \quad x_i' \in [0,1]
$$



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
df = pd.read_csv('wine_data.csv', header = None, usecols=[0,1,2])
df.columns = ['Class label', 'Alcohol', 'Malic acid' ]
```

```python
df 
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict = {1:'red', 2:'green', 3:'blue'}
sns.scatterplot(x=df['Alcohol'], y=df['Malic acid'], hue=df['Class label'], palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1), df['Class label'], test_size=0.3, random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# fit the scaler to train set, it will learn from parameter
scaler.fit(X_train)

# transform train and test set
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig,(ax1, ax2)= plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c=y_train)
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'], c=y_train)
ax2.set_title('After Scaling')
plt.show()
```

```python
fig,(ax1, ax2)= plt.subplots(ncols=2, figsize=(12,5))
#Before Scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Malic acid'], ax=ax1)
sns.kdeplot(X_train['Alcohol'], ax=ax1)
#After Scaling
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
ax2.set_title('After Scaling')
plt.show()
```

### 2. Mean Normalization

$$
x_i' = \frac{x_i - \bar{x}}{x_{\max} - x_{\min}}, \quad 
x_i' \in \left[-1,1\right], \quad
\text{data centered at 0 and scaled to the feature range}
$$


### 3. MaxAbsScaling

$$
x_i' = \frac{x_i}{|x_{\max}|}, \quad x_i' \in [-1, 1]
$$
$$\text{Class in sklearn = MaxAbsScaling}$$
$$\text{used for sparse data(having many zero)}$$

### 4. Robust Scaling

$$
x_i' = \frac{x_i - \text{median}(x)}{\text{IQR}(x)}, \quad 
\text{where } \text{IQR}(x) = Q_3 - Q_1
$$
