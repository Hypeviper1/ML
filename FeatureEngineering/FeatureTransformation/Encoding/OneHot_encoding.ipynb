## OneHot Encoding

use to (n-1) for n column

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

### 1. OneHot Encoding using as Pandas

```python
pd.get_dummies(df, columns=['fuel','owner'], dtype=int)
```

### 1. K-1 OneHot Encoding using as Pandas

```python
pd.get_dummies(df, columns=['fuel','owner'], drop_first=True, dtype=int)
```

### 3. OneHotEncoding using sklearn

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1], test_size = 0.3, random_state = 0)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
x_train_new= ohe.fit_transform(x_train[['fuel','owner']])
```

### 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts<= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'), dtype = int).sample(5)
```
